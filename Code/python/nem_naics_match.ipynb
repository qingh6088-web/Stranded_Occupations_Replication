```python
import pandas as pd
import os
import glob
```

```python
# File paths
naics_file_path = "Your path/Stranded_Occupations_Replication/Data/raw/NAICS_nem_code.xlsx"
matrix_file_path = "Your path/strand_labor/Stranded_Occupations_Replication/Data/raw/matrix.xlsx"

# Read NAICS_nem_code file
print("Reading NAICS_nem_code.xlsx...")
try:
    naics_df = pd.read_excel(naics_file_path)
    print(f"Successfully read NAICS file: {len(naics_df)} rows")
    print(f"Columns: {naics_df.columns.tolist()}")
    print("First 5 rows:")
    print(naics_df.head())
except Exception as e:
    print(f"Error reading NAICS file: {e}")
    exit()

# Read matrix file
print("\nReading matrix.xlsx...")
try:
    matrix_df = pd.read_excel(matrix_file_path)
    print(f"Successfully read matrix file: {len(matrix_df)} rows")
    print(f"Columns: {matrix_df.columns.tolist()}")
    print("First 5 rows:")
    print(matrix_df.head())
except Exception as e:
    print(f"Error reading matrix file: {e}")
    exit()

# Check required columns
required_naics_cols = ['NAICS2022', 'NAICS2022title', 'nem_indcode']
required_matrix_cols = ['nem_indcode', 'Occupation code', 'Occupation title']

missing_naics_cols = [col for col in required_naics_cols if col not in naics_df.columns]
missing_matrix_cols = [col for col in required_matrix_cols if col not in matrix_df.columns]

if missing_naics_cols:
    print(f"Missing columns in NAICS file: {missing_naics_cols}")
    exit()

if missing_matrix_cols:
    print(f"Missing columns in matrix file: {missing_matrix_cols}")
    exit()

# Ensure consistent data type for nem_indcode
naics_df['nem_indcode'] = naics_df['nem_indcode'].astype(str)
matrix_df['nem_indcode'] = matrix_df['nem_indcode'].astype(str)

# Merge data
print("\nMerging data...")
merged_df = pd.merge(naics_df, matrix_df, on='nem_indcode', how='left')

print(f"Merged data: {len(merged_df)} rows")

# Check unmatched records
unmatched_naics = naics_df[~naics_df['nem_indcode'].isin(matrix_df['nem_indcode'])]
if len(unmatched_naics) > 0:
    print(f"\n{len(unmatched_naics)} NAICS industries not matched in matrix:")
    for _, row in unmatched_naics.iterrows():
        print(f"  {row['NAICS2022']} - {row['NAICS2022title']} (nem_indcode: {row['nem_indcode']})")

# Save results
output_dir = "Your path/Stranded_Occupations_Replication/Data/raw"
os.makedirs(output_dir, exist_ok=True)

output_file = os.path.join(output_dir, "naics_nem_new.xlsx")
merged_df.to_excel(output_file, index=False)

print(f"\nMerge completed!")
print(f"Results saved to: {output_file}")

# Display summary statistics
print(f"\nMerged data summary:")
print(f"Total records: {len(merged_df)}")
print(f"Unique NAICS industries: {merged_df['NAICS2022'].nunique()}")
print(f"Unique occupations: {merged_df['Occupation code'].nunique()}")

# Display first few rows
print(f"\nFirst 5 rows of merged data:")
print(merged_df.head())
```
